# Asset Pricing

Asset prices are the value of claims to future uncertain payments.

$$
p = E(mx)
$$


# Notation
Price: $p_t$\
Payoff: $x_t$\
Dividends: $D_t$\
Return: $r_t$\
**Stochastic discount factor (SDF)**: $m_t$


# 1. Fundamental Relationship

Where:

        The relationsihp between prices, payoffs, and dividends:
$$
x_{t+1} = p_{t+1} + D_{t+1}
$$

        The relationsihp between prices, payoffs, and returns:
$$
R_{t+1} = \frac{x_{t+1}}{p_t}\tag{2.2}
$$ 
**Remark**: All asset pricing models focus particularly on returns rather than prices cuz: 
- Returns are scale-free
- Stationary

# 2. Fundamental Asset Pricing Equation:



Since the financial market is both dynamic and uncertain:
$$
p_t = E_t(m_{t+1}x_{t+1})\tag{1}
$$

Where $p_t$ is the price of the asset at time t, $E_t$ is the conditional expectation at time $t$, $m_{t+1}$ is the **stochastic discount factor (SDF)** and $x_{t+1}$ is the asset's payoff. This is called the discount factor approach

Rewrite the equation in return form:
$$
\frac{p_t}{p_t} = E_t(m_{t+1}\frac{x_{t+1}}{p_t})\tag{2.1}
$$

cuz (2.2), (2.1) is:

$$
1 = E_t(m_{t+1}R_{t+1})\tag{3}
$$
(3) this the **fundamental asset return equation**

# 3. Risky/Risk-free Return

**Uncertainty** in payoffs make most assets **risky**.\
We donote the expected returns of a risky asset $i$ as $E_t(R_{t+1}^i)$\
So, the **risky returns** following (3):
$$
1 = E_t(m_{t+1}R_{t+1}^i)
$$

No unvertainty in future will make assets **risk-free**.\
A **risk-free return** is denoted as $R^f$. So the risk-free return is :
$$
1 = E_t(m_{t+1}R^f) = E_t(m_{t+1})R^f
$$
So:
$$
R^f = \frac{1}{E_t(m_{t+1})}
$$
e.g.:
US Treasury Bills (T-bills, T-notes, T-bonds)

Excess return is the premium as investor expects to get for holding risky asset.\
So, holding risky asset is compensated by a risky return\
Excess returns is expressed as the difference between expected returns of the risky asset and a risk-free assets:
$$
Excess Return = E_t(R_{t+1}^i) - R^f
$$
